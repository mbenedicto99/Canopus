```python
import zipfile
import os

zip_file_path = '/content/smallpdf-convert-20260407-143806.zip'
extraction_dir = '/content/extracted_smallpdf'

# Create the extraction directory if it doesn't exist
os.makedirs(extraction_dir, exist_ok=True)

# Open the zip file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    print(f"Conteúdo do arquivo zip '{zip_file_path}':")
    for member in zip_ref.namelist():
        print(f" - {member}")

    # Extract all the contents into the specified directory
    zip_ref.extractall(extraction_dir)
    print(f"\nTodos os arquivos foram extraídos para: {extraction_dir}")

print("\nArquivos extraídos:")
for root, dirs, files in os.walk(extraction_dir):
    for name in files:
        print(os.path.join(root, name))
    for name in dirs:
        print(os.path.join(root, name))
```

```python
import os
from PIL import Image
import pytesseract

# Define a variável de ambiente TESSDATA_PREFIX
os.environ['TESSDATA_PREFIX'] = '/usr/share/tesseract-ocr/4.00/tessdata'

# Certifique-se de que a variável extraction_dir está definida
# Se não estiver, defina-a aqui novamente. Neste caso, ela deve estar definida
# pela execução das células anteriores.
# extraction_dir = '/content/extracted_smallpdf'

extracted_texts_tesseract = {}

# Obtém a lista de todos os arquivos de imagem no diretório de extração
image_files = [f for f in os.listdir(extraction_dir) if f.lower().endswith(('.jpg', '.png', '.jpeg'))]
image_files.sort() # Ordena para uma ordem consistente

print(f"Iniciando o processamento de {len(image_files)} imagens com Tesseract OCR.")

for image_filename in image_files:
    file_path = os.path.join(extraction_dir, image_filename)

    print(f"\nProcessando imagem: {image_filename}")

    try:
        # Abre a imagem
        img = Image.open(file_path)

        # Realiza o OCR usando pytesseract
        # 'por' para português. Se o texto for em inglês, use 'eng'.
        text = pytesseract.image_to_string(img, lang='por')
        extracted_texts_tesseract[image_filename] = text

        print(f"Texto extraído de {image_filename} (primeiros 500 caracteres):\n{text[:500]}...")
    except Exception as e:
        print(f"Erro ao processar a imagem {image_filename} com Tesseract: {e}")

print("\n--- Sumário dos textos extraídos com Tesseract ---")
for filename, text in extracted_texts_tesseract.items():
    print(f"\nArquivo: {filename}\nTexto (primeiros 200 caracteres):\n{text[:200]}...")
```

```python
import os
from PIL import Image
import pytesseract # Adicionada a importação aqui

# Define a variável de ambiente TESSDATA_PREFIX
os.environ['TESSDATA_PREFIX'] = '/usr/share/tesseract-ocr/4.00/tessdata'

extraction_dir = '/content/extracted_smallpdf'
image_files = [f for f in os.listdir(extraction_dir) if f.lower().endswith(('.jpg', '.png', '.jpeg'))]
image_files.sort()

if image_files:
    # Processa a primeira imagem como exemplo
    sample_image_path = os.path.join(extraction_dir, image_files[0])
    print(f"\nProcessando a primeira imagem com Tesseract: {image_files[0]}")

    try:
        # Abre a imagem
        img = Image.open(sample_image_path)

        # Realiza o OCR usando pytesseract
        text = pytesseract.image_to_string(img, lang='por') # 'por' para português
        print(f"Texto extraído de {image_files[0]} (primeiros 500 caracteres):\n{text[:500]}...")
    except Exception as e:
        print(f"Erro ao processar a imagem {image_files[0]} com Tesseract: {e}")
else:
    print("Nenhuma imagem encontrada para processar.")
```

Este é um exemplo de como usar o `pytesseract` em uma única imagem. Para processar todas as imagens, você pode adaptar o loop que eu havia criado anteriormente, substituindo a chamada à Cloud Vision API pela função `pytesseract.image_to_string()`.

```python
output_filename = '/content/extracted_texts_from_images.txt'

combined_text = ""
for filename, text in extracted_texts_tesseract.items():
    combined_text += f"--- Texto de {filename} ---\n"
    combined_text += text
    combined_text += "\n\n"

with open(output_filename, 'w', encoding='utf-8') as f:
    f.write(combined_text)

print(f"Todos os textos extraídos foram salvos em: {output_filename}")

# Você pode visualizar o conteúdo do arquivo com este comando (descomente a linha abaixo):
# !cat {output_filename}
```

```python
import re

cleaned_texts = {}

for filename, text in extracted_texts_tesseract.items():
    # Convert to lowercase
    text_lower = text.lower()
    # Remove special characters (keep alphanumeric and spaces)
    cleaned_text = re.sub(r'[^a-z0-9\s]', '', text_lower)
    cleaned_texts[filename] = cleaned_text

print("--- Sumário dos textos limpos ---")
for filename, text in cleaned_texts.items():
    print(f"\nArquivo: {filename}\nTexto limpo (primeiros 200 caracteres):\n{text[:200]}...")
```

```python
output_cleaned_filename = '/content/cleaned_extracted_texts.txt'

combined_cleaned_text = ""
for filename, text in cleaned_texts.items():
    combined_cleaned_text += f"--- Texto Limpo de {filename} ---\n"
    combined_cleaned_text += text
    combined_cleaned_text += "\n\n"

with open(output_cleaned_filename, 'w', encoding='utf-8') as f:
    f.write(combined_cleaned_text)

print(f"Todos os textos limpos foram salvos em: {output_cleaned_filename}")

# Você pode visualizar o conteúdo do arquivo com este comando (descomente a linha abaixo):
# !cat {output_cleaned_filename}
```
